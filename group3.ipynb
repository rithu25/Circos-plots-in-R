```
import pandas as pd
```

```
newdf3 = pd.read_csv('Group3new.txt', sep="\t")
```

```
newdf3
```

```
newg3 = pd.read_csv("group3newciiider.csv")
```

```
newe3 = pd.read_excel("enrich3.xlsx")
```

```
newe3['Transcription Factor Name'].nunique()
```

```
newg3.rename(columns={'Gene Name':'Genename'}, inplace=True)
```

```
newg3[['Chrom','Start','End','Genename']] = newg3.Genename.str.split("_",expand=True,)
```

```
newg3
```

```
newmerged_df = newg3.merge(newe3, how = 'inner', on = ["Transcription Factor ID","Transcription Factor Name"])
```

```
newmerged_df
```

```
newg3 = newmerged_df[["Chrom","Start Position",'End Position','Genename','Transcription Factor Name','Transcription Factor ID']].copy()
```

```
newmeh3 = newdf3.merge(newg3,how='left', left_on='Genename', right_on='Genename')
```

```
newmeh3
```

```
###Add start and end positions
newmeh3['Start'] = newmeh3['Start'] + newmeh3['Start Position'] 
```

```
###Add start and end positions
newmeh3['End'] = newmeh3['End'] + newmeh3['End Position'] 
```

```
newmeh3.head()
```

```
newmeh3 = newmeh3[["Chrom_x","Start",'End','Genename','Transcription Factor Name']].copy()
```

```
newmeh3.rename(columns={'Chrom_x':'Chrom'}, inplace=True)
```

```
newmeh3.to_csv('group3ciiider.csv', index = False)
```
